## Chess rules
Board
Piece color:
* 黑 = black
* 红 = red
Pieces:
* 帥 - seoi6 = general red
* 將 - zoeng1 = general black
* 仕 - si6 = advisor red
* 士 - si6 = advisor black
* 相 - soeng4 = elephant red
* 象 - zoeng6 = elephant black
* 車 - ce1 = rook red & black
* 馬 - maa5 = horese red & black
* 炮 - paau4 = cannon red & black
* 兵 - bing1 = soldier black
* 卒 - zuk1 = soldier red
Besides elephants (xiang 象), Chinese chess also includes generals (jiang 将) , soldiers (zu 卒), horses (ma 马), chariots (ju or che 車) , guards (advisors, counselors) (shi 士), canons (pao 炮), palaces (gong 宫) and even a river (he 河).



```python
# C 1.
class Piece:
    def __init__(self, name, color):
        self.name = name
        self.color = color

# C 2.
class Board:
    def __init__(self):
        self.board = self.create_board()

    @staticmethod
    def create_board():
        board = [['' for _ in range(9)] for _ in range(10)]
        pieces = ['Chariot', 'Horse', 'Elephant', 'Advisor', 'King', 'Advisor', 'Elephant', 'Horse', 'Chariot']
        colors = ['Red', 'Black']
        for i in range(2):
            for j in range(9):
                board[i*9][j] = Piece(pieces[j], colors[i])
                board[i*9+1][j] = Piece('Soldier', colors[i]) if j % 2 == 0 else ''
        return board
```
